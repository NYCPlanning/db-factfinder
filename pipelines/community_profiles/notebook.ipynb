```python
variable_mapping = []
for i in a.split(','):
    i = i.strip().replace('\n', '')
    var_type = i[8:10].strip()
    pff_variable = i[38:].split("'")[0].strip().lower()
    geotype=i[38:].split("and geotype='")[1].split("'")[0].strip().lower()
    target=i[38:].split('as')[1].strip()
    variable_mapping.append(
        dict(
             var_type=var_type, 
             pff_variable=pff_variable, 
             geotype=geotype, 
             target=target
        ))
```

```python
df['column_mapping'] = df.apply(lambda x: {x['var_type']:x['target']}, axis=1)
geotype_translate = {
    'puma2010':'cd',
    'city2010': 'city',
    'boro2010': 'borough'
}
df['geotype'] = df.apply(lambda x: geotype_translate.get(x['geotype']), axis=1)
```

```python
def agg(df):
    out={}
    for i in df['column_mapping'].to_list():
        out.update(i)
    return out
dff = df.groupby(['pff_variable', 'geotype']).apply(agg)
```
